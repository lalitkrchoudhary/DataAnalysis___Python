### Pandas-DataFrame AND Series
Pandas is a poweful data  manupulation library in Python, widely used for data analysis and data cleaning. It provides two primary data structure: Series and DataFrame.A Series is a one-dimension array-like object, while a DataFrame is two dimensional, size mutable, and potentially heterogeneous tabular data structure with labeled axes(row and colums).

```python
import pandas as pd
data=[1,2,3,4,5]
series=pd.Series(data)
print("Series \n",series) # first one is the index and second column is data "it have defalut index when using list"
```

```python
#Create a Series using dictionary
data={"a":1,"b":2,"c":4}
series_dict=pd.Series(data)
print(series_dict)
```

```python
data=[10,20,30]
ind={'a','b','c'}
pd.Series(data,index=ind)
```

```python
#DataFrame(It can have multiple row and columns)

#Create a dataframe from a dictionary of list

data={
    'Name':['Krish','John','Jack'],
    'Age' :[25,30,45],
    'City' :['Bangalore',"New York",'Florida']
}

df=pd.DataFrame(data)
print(df)
```

### Seaborn
Seaborn is a Python visualization library based on Matplotlib that provides a high-level for drawing attractive and informative statistical graphics. Seaborn helps in creating complex visualizations with just a few lines of code.

```python
import matplotlib.pyplot as plt
```

```python
import seaborn as sns
```

```python
#Basic Plotting with seaborn
tips=sns.load_dataset('tips')
tips
```

```python
import matplotlib.pyplot as plt

sns.scatterplot(x='total_bill',y='tip',data=tips)
plt.title('Scatter Plot of Total Bill vs Tip')
plt.show()
```

```python
#Line PLot
sns.lineplot(x='size',y='total_bill',data=tips)
plt.title("Line plot of Total BIll by size")
plt.show()
```

```python
#Categorical Plot
#Bar Plot
sns.barplot(x='day',y='total_bill',data=tips)
plt.title('Bar PLot of Total Bill By Day')
plt.show()
```

```python
#Box plot
sns.boxplot(x='day',y='total_bill',data=tips)
plt.title('Bar plot of TOtal Bill vs days')
plt.show()
```

```python
#Violine plot
sns.violinplot(x='day',y='total_bill',data=tips)
plt.title('Violine plot of total bill vs days')
plt.show()
```

```python
tips
```

```python
#Histograms
sns.histplot(tips['total_bill'],bins=10,kde=True)
```

```python
#KDE PLOT
sns.kdeplot(tips['total_bill'],fill=True)
```

```python
#Pairplot
sns.pairplot(tips)
```

```python
#Heatmap work on numerical data only
corr=tips[['total_bill','tip','size']].corr()
corr
```

```python
sns.heatmap(corr,annot=True,cmap='coolwarm')
```
